```python
import subprocess
import scipy as sp
import rasterio
import numpy as np
import meshio
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('C:\\Users\\Andrea\\Documents\\Flow\\Optimization\\headData\\averageInTime\\wellHeadsPlain.csv')
wellLoc = df
wellLoc['Bottom'] = wellLoc['quota[m]']-wellLoc['depth']

def flowSimulation(P):
    #model parameters
    
    myfile = open(r"C:\Users\Andrea\Documents\Flow\OpenGeoSys\ogs6test\Po1.prj", "rt")
    contents = myfile.readlines()
    myfile.close()
    
    wordX = 'permeability'
    j=0
    for i in range(len(contents)):
        if wordX in contents[i]:
            value = contents[i+2].split('<value>').pop().split('</value>')[0]
            if j>0:
                K = 10**P[j-1]
                value2= str(K)
                contents[i+2] = contents[i+2].replace(value, value2)
                permName = 'K%1d' % j 
                print(permName, value2)
            j += 1
    with open(r"C:\Users\Andrea\Documents\Flow\OpenGeoSys\ogs6test\Po2.prj", 'w') as file:
        file.writelines( contents )
    
    
    subprocess.check_call([r"C:\Users\Andrea\Documents\Flow\OpenGeoSys\ogs-6.4.3-16-g9a7564950-Windows-10.0.22000-python-3.10.7-utils\bin\ogs.exe", r"C:\Users\Andrea\Documents\Flow\OpenGeoSys\ogs6test\Po2.prj"])

    mesh = meshio.read(r"C:\Users\Andrea\Documents\Flow\OpenGeoSys\ogs6test\Po_ts_1_t_1.000000.vtu")
    data = mesh.points
    sample = np.array([wellLoc.longitude,wellLoc.latitude,wellLoc.Bottom]).T
    kdtree=sp.spatial.KDTree(data)
    dist,points=kdtree.query(sample,1)
    simulatedH = (mesh.point_data['pressure'][points].T)/9810+sample[:,2]
    fobj = (sum(((simulatedH-np.array(wellLoc['head[m]']))**2)**(1/2)))/len(wellLoc['head[m]'])
    print('RMSE: ',fobj)
    return fobj
```

```python
P = [-8,-9,-9,-10,-10,-11]
```

```python
flowSimulation(P)
```

```python
#define hyperparameters
F=0.5
CR = 0.5
NP = 60
bounds = [(-7, -9),(-8, -10),(-8, -10),(-9, -13),(-9, -13),(-11, -15)]
maxIter = 200
tollerance = 1e-9
```

```python
result = sp.optimize.differential_evolution(
    flowSimulation,
    bounds=bounds,
    maxiter=maxIter,
    popsize=NP,
    tol=tollerance, 
    mutation=F,
    disp=True,
    recombination = CR,
    polish=False,
    #updating='immediate',
    #workers=1
)
print(result)
```
